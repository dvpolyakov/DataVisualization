```python
import pandas as pd
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
df = pd.read_csv('eucountrydatawb_csv.csv')
df.sort_values('population in millions', inplace=True, ascending=False)
df.head()
```

```python
populations = df['population in millions'].values
countries = df['jurisdiction'].values
populations = list(map(float, populations))
```

```python
from squarify import Squarify
```

```python
im = np.array(Image.open('img.jpg'), dtype=np.uint8)
x = 0
y = 0
h = im.shape[0]
w = im.shape[1]

s = Squarify()
res = s.squarify(populations, x, y, w, h)
```

```python
# draw results
fig,ax = plt.subplots(1)
ax.imshow(im)

for counter, elem in enumerate(res):
    color = np.random.rand(3,)
    rect = patches.Rectangle((elem['x'], elem['y']),
                              elem['dx'], elem['dy'],
                              linewidth=2, fill=True,
                              edgecolor=color,
                              facecolor=color,
                              label=countries[counter])
    ax.add_patch(rect)
    plt.legend(bbox_to_anchor = (1,1.03))
    
plt.title("Population of European countries")
plt.show()
```
